```python
import pandas as pd
import numpy as np
```

```python
with open("srr2prot.txt", 'r') as f:
    lsq = list(map(lambda x: x.strip('\n').split('\t'), f.readlines()))
dct = dict()
for element in lsq:
    dct[element[0]] = (element[1], element[2])
with open("fastqdump_log.txt", 'r') as f:
    ls = list(map(lambda x: x.rstrip('\n').lstrip("Written ").split(" spots for "), f.readlines()))
lsa = dict(map(lambda x: [x[1], int(x[0])], ls))
dct2 = {x : 0 for x in dct.keys()}
for key in list(dct.keys()):
    start, end = int(dct[key][0].strip("SRR")), int(dct[key][1].strip("SRR"))
    for i in range(start, end + 1):
        dct2[key] += lsa["SRR" + str(i)]

rds = dct2
print(rds.keys())
```

```python
replicas = [9, 3, 3, 3, 3, 3, 3, 3, 3, 2, 3, 3, 3, 3, 
           3, 3, 3, 3, 3, 3, 3, 3, 2, 3, 3]
repls = {x: y for x, y in zip(list(rds.keys()), replicas)}
```

```python
for key in rds.keys():
    rds[key] = rds[key], repls[key], dct[key][0], dct[key][1]
```

```python
metadata = pd.DataFrame(rds).transpose().rename(columns={0:"reads_mln", 1:"replicas", 2: "start", 3:"end"})
metadata["prot"] = metadata.index
metadata
```

```python
fqc = pd.read_csv("multiqc_data/mqc_bowtie2_pe_plot_1.txt", sep='\t')
fqc = fqc.assign(Sample = lambda x: x.Sample.str.strip("mapped_reads | log_hisat_"))
fqc
```

```python
align = {x: [] for x in metadata.prot}
for p in metadata.prot:
    start = dct[p][0]
    end = dct[p][1]
    tmp_df = fqc[(fqc.Sample >= start) & (fqc.Sample <= end)]
    align[p] = list(map(lambda x: round(sum(tmp_df[x])), tmp_df.columns[1:]))
```

```python
aligning = pd.DataFrame(align).transpose().rename(columns={
    0:"PE mapped uniquely",
    1:"PE mapped discordantly uniquely",
    2:"PE one mate mapped uniquely",
    3:"PE multimapped",
    4:"PE one mate multimapped",
    5:"PE neither mate aligned"
})
aligning["prot"] = aligning.index
```

```python
anot = pd.read_csv("annot_metrics.txt", sep=' ', header = None)
anot = anot.iloc[:, 1:3].rename(columns={1:"n_anot", 2:"SRR"}).assign(
    SRR = lambda x: x.SRR.str.lstrip("contacts/").str.rstrip(".bed"))
anot
```

```python
annot = {x: [] for x in metadata.prot}
for p in metadata.prot:
    start = dct[p][0]
    end = dct[p][1]
    tmp_df = anot[(anot.SRR >= start) & (anot.SRR <= end)]
    annot[p] = sum(tmp_df.n_anot)
annot
```

```python
b = pd.merge(metadata, aligning)
b
```

```python
1 - b["PE neither mate aligned"] / b.reads_mln
```

```python
preannot = pd.read_csv("preannot_counts.txt", header = None, sep=" ").rename(columns={0: "n", 1:"SRR_id"})
annot = pd.read_csv("annot_counts.txt", header = None, sep=" ").rename(columns={0: "m", 1:"SRR_id"})
annot_stats = pd.merge(annot, preannot)
```

```python
percent_annot = {}
for p in metadata.prot:
    start = dct[p][0]
    end = dct[p][1]
    tmp_df = annot_stats[(annot_stats.SRR_id >= start) & (annot_stats.SRR_id <= end)]
    percent_annot[p] = round(1/(sum(tmp_df.n)/sum(tmp_df.m)),3)
percent_annot
```
